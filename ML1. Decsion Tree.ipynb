```python
# Machinelearning understands only numbers so importing Scikit learn label encoder
from sklearn.preprocessing import LabelEncoder
```

```python
# Machine learning algorithm for classification
from sklearn import tree
```

```python
import pandas as pd
df = pd.read_csv("/salaries.csv")
df.head()
```

```python
#to remove/drop one column
inputs = df.drop('salary_more_then_100k',axis='columns')

#assign that dropped column to a dataframe named "Target"
Target = df['salary_more_then_100k']
```

```python
inputs
```

```python
Target
```

```python
comp = LabelEncoder()
job = LabelEncoder()
degree = LabelEncoder()
```

```python
#creates a seperate column for company,job and degree with labled numbers

inputs['company_n'] = comp.fit_transform(inputs['company'])
inputs['job_n'] = job.fit_transform(inputs['job'])
inputs['degree_n'] = degree.fit_transform(inputs['degree'])
inputs.head()
```

```python
inputs_n = inputs.drop(['company','job','degree'],axis='columns')
inputs_n
```

```python
model = tree.DecisionTreeClassifier()
model1 = tree.ExtraTreeClassifier()
```

```python
 model.fit(inputs_n,Target)

 model1.fit(inputs_n,Target)
```

```python
# to predict the accuracy of the model
#since its a small dataset accuracy is 100% = 1

model.score(inputs_n,Target)
```

```python
# to predict the accuracy of the model1
#since its a small dataset accuracy is 100% = 1

model1.score(inputs_n,Target)
```
