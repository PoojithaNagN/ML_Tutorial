```python
file_path = '/content/drive/MyDrive/Data /Melborne_housing_prices.csv'
```

```python
df = pd.read_csv(file_path)
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df = df[['Suburb','Rooms','Type','Price','SellerG','Regionname','Propertycount','Distance','CouncilArea']]
```

```python
#shows the sum of NA values in each column
df.isnull().sum()
```

```python
#drops NA values in the column
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df = pd.get_dummies(df,drop_first=True)
```

```python
X = df.drop('Price',axis=1)
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
