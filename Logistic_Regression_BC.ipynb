# **Credit card analysis using Logistic Regression! - Binary Classification**

Logistic Regression is one of the technique used for classification.

Predicted value from logistic regression is categorical. Answers questions like:


1.  Binary Classification : Has a credit card with the bank? Yes/No



```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
file_path = '/content/drive/MyDrive/Data /Bank_Churn.csv'
```

```python
df = pd.read_csv(file_path)
df.head()
```

```python
df = df[['CreditScore','HasCrCard']]
df.head()
```

```python
plt.scatter(df['CreditScore'], df['HasCrCard']) # Changed plot to plt and passed the columns as series
plt.xlabel('CreditScore')  # Added labels for better visualization
plt.ylabel('Has Credit Card')
plt.title('Scatter Plot of CreditScore vs. Has Credit Card')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

X_train, X_test, Y_train, Y_test = train_test_split(df[['CreditScore']], df['HasCrCard'], test_size=0.1)
```

```python
X_test
```

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
model.predict(X_test)
```

```python
#the accuracy is 67%

model.score(X_test, Y_test)
```
